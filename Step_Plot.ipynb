Step Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.grid()
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.grid()
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.step(x,y,color= "r" ,marker ="o" )
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.grid()
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.step(x,y,color= "r" ,marker ="o", ms=10 , mfc = "g", label ="Python" )
plt.legend(loc =2)
plt.show()
```
